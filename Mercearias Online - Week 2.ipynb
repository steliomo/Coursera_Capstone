```python
address = 'Maputo, Mozambique'
geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

### Define the url to get more groceries

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query,LIMIT)
results = requests.get(url).json()

# assign relevant part of JSON to venues
groceries = results['response']['venues']

# tranform groceries into a dataframe
grocery_df = json_normalize(groceries)
grocery_df
```

### Data cleaning

```python
grocery_df.drop(['id', 'categories', 'referralId', 'hasPerk', 'location.labeledLatLngs', 'location.labeledLatLngs', 'location.distance', 'location.cc', 'location.state', 'location.country', 'location.address'], axis=1, inplace=True)
```

```python
grocery_df.head()
grocery_df.rename(columns= {"location.lat":"lat", "location.lng": "lng", "location.city":"city", "location.formattedAddress": "address"}, inplace=True)
```

```python
groceries = pd.concat([df, grocery_df])
groceries.to_excel("grocery_rollout_with_foursquare.xlsx", engine='xlsxwriter')
```
